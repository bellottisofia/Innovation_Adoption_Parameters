```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.spatial.distance import pdist, squareform
from sklearn.impute import SimpleImputer
```

```python
opinion_germany = pd.read_csv('/content/matrix_Germany (4).csv')
```

```python
opinion_germany.head()
```

```python

popolazioni = opinion_germany['popolazione'].values
opinioni = opinion_germany.drop(columns=['popolazione']).values


# 4. distance matrix
distanze = squareform(pdist(opinioni, metric='euclidean'))

# 5. sigma
sigma_media_germany = np.mean(distanze[np.triu_indices_from(distanze, k=1)])
sigma_mediana_germany = np.median(distanze[np.triu_indices_from(distanze, k=1)])

print("Mean Sigma:", sigma_media_germany)
print("Median Sigma:", sigma_mediana_germany)
# 3. construction of the similarity matrix W using Gaussian Kernel
W_germany1 = np.exp(-distanze**2 / (2 * sigma_media_germany**2))
W_germany2 = np.exp(-distanze**2 / (2 * sigma_mediana_germany**2))
```

We normalize each similarity matrix by rows, in order to obtain stochastic matrices.

```python

row_sums1 = W_germany1.sum(axis=1, keepdims=True)
W_germany1 = W_germany1 / row_sums1
row_sums2 = W_germany2.sum(axis=1, keepdims=True)
W_germany2 = W_germany2 / row_sums2
```

```python

similarity_matrix_germany1 = pd.DataFrame(W_germany1, index=popolazioni, columns=popolazioni)
similarity_matrix_germany1.to_csv("similarity_germany_mean.csv", index=True)
similarity_matrix_germany2 = pd.DataFrame(W_germany2, index=popolazioni, columns=popolazioni)
similarity_matrix_germany2.to_csv("similarity_germany_median.csv", index=True)
```
